```python
#importing the relevant libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("C:/Users/shubh/Documents/Data Viz Project/2019_Data_Professional_Salary_Survey_Responses.xlsx",
                  sheet_name="Salary Survey") 
print (df)
```

```python
# Bar Graph for each year showcasing the count of records
```

```python
yearCount = df.groupby("Survey Year")["Survey Year"].count()
yearCount.plot.barh(figsize=(10,5), color="#ff6361")
plt.ylabel('Survey Year')
plt.xlabel('Count')
plt.title('Count of records for each year')
plt.tight_layout()

#We can comprehend that year 2018 has the maximum records hence, maximum number of survey takers
```

```python
#to calculate the ratios and percentages of each survey year
print("The following are the percentages of records for each year")
yearTotal = df["Survey Year"].count()
(df['Survey Year'].value_counts()/yearTotal)*100
```

```python
#Bar Graph for Exploring the SalaryUSD Column using a vertical histogram
salary=df["SalaryUSD"].hist(bins=40, figsize=(20,20), color="thistle", grid=False, edgecolor='black', linewidth=1.5)
salary.set_xlabel("Salary Brackets", labelpad=20, weight='bold', size=20)
salary.set_ylabel("Count", labelpad=20, weight='bold', size=20)
salary.tick_params(axis="x", labelsize=15)
salary.tick_params(axis="y", labelsize=15)
salary

#We can comprehend that majority of people who took the survey had their salaries between the range 0 to 200,000 USD
```

```python
#Exploring the count of participants VS countries they belong to
```

```python
countryCount = df.groupby("Country")["Country"].count()
countryCount.plot.barh(figsize=(20,20), color="steelblue", edgecolor='black', linewidth=1.0)
plt.ylabel('Country Name', fontsize = 17, weight ="bold")
plt.xlabel('Count of people taking the survey', fontsize = 17, weight="bold")
plt.title('Count of People vs Country', fontsize = 19, weight ="bold")
plt.tick_params(axis="x", labelsize=12)
plt.tick_params(axis="y", labelsize=12)
plt.tight_layout()

#We can comprehend that USA, UK, Canada and Australia had the most number of people participating in this survey
```

```python
print("Count of survey participants from each country")
df["Country"].value_counts()
```

```python
#Exploring the column - Employment Status using a Horizontal Bar Graph
ESCount = df.groupby("EmploymentStatus")["EmploymentStatus"].count()
ESCount.plot.barh(figsize=(10,5), color="#bc5090", edgecolor='black', linewidth=1.0)
plt.ylabel('Employment Status', fontsize = 12, weight ="bold")
plt.xlabel('Count', fontsize = 12, weight ="bold")
plt.title('Count vs Employment Status', fontsize = 14, weight ="bold")
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.tight_layout()

#We can comprehend that majority of the people who took the survey are/ were full time employees and the least are part time employees.
```

```python
#Exploring the column - Education using a simple horizontal bar graph
EduCount = df.groupby("Education")["Education"].count()
EduCount.plot.barh(figsize=(10,5), color="#58508d", edgecolor='black', linewidth=1.0)
plt.ylabel('Education Qualification', fontsize = 12, weight ="bold")
plt.xlabel('Count', fontsize = 12, weight ="bold")
plt.title('Count vs Education Qualification', fontsize = 14, weight ="bold")
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.tight_layout()

#We can comprehend that majority of the people who took the survey have a Bachelor's degree, followed by a Master's degree
```

```python
#Exploring the column - Employment Sector using a horizontal bar graph
EmpSecCount = df.groupby("EmploymentSector")["EmploymentSector"].count()
EmpSecCount.plot.barh(figsize=(10,5), color="#003f5c", edgecolor='black', linewidth=1.0)
plt.ylabel('Employment Sector', fontsize = 12, weight ="bold")
plt.xlabel('Count', fontsize = 12, weight ="bold")
plt.title('Count vs Employment Sector', fontsize = 14, weight ="bold")
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.tight_layout()

#We can comprehend that most of the people belong to the Private business sector. 
```

```python
#Exploring the column - Gender using a horizontal bar graph 
GCount = df.groupby("Gender")["Gender"].count()
GCount.plot.barh(figsize=(10,10), color="#ffa600", edgecolor='black', linewidth=1.0)
plt.ylabel('Gender Category', fontsize = 12, weight ="bold")
plt.xlabel('Count', fontsize = 12, weight ="bold")
plt.title('Count VS Gender Category', fontsize = 14, weight ="bold")
plt.tick_params(axis="x", labelsize=10)
plt.tick_params(axis="y", labelsize=10)
plt.tight_layout()

#W can comprehend that most of the people who took the survey are males. 
#we plan to consider only 5 gender categories for future work, these are - Male, Female, Not Asked, Prefer not to say, Third Gender
```

```python
#Exploring the relation between Country and HoursWorkedPerWeek
country_hours = df.groupby("Country")["HoursWorkedPerWeek"].mean()
country_hours.plot.barh(figsize=(20,20), color="c", edgecolor='black', linewidth=1.0)
plt.xlabel('Average Hours Worked Per Week', fontsize = 17, weight ="bold")
plt.ylabel('Country', fontsize = 17, weight ="bold")
plt.title('Country vs Average Hours Worked', fontsize = 19, weight ="bold")
plt.tick_params(axis="x", labelsize=12)
plt.tick_params(axis="y", labelsize=12)
plt.tight_layout()

#It seems like the average number of hours worked across almost all the countries is between 30 to 40. Albania is the only exception with 160 hours per week.This is because there is only one record from Albania.  
```
